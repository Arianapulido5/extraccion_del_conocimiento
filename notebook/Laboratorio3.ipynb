```python
import pandas as pd
```

```python
import random
```

```python
from datetime import datetime, timedelta
```

```python
influencers = [
    "cancun_vibes", "playadelcarmen_life", "rivieramaya_travel",
    "cancun_foodie", "mexico_caribbean", "tulum_lifestyle",
    "cancun_sunsets", "yucatan_explorer", "caribemexicano",
    "cancun_turismo"
]

tipos_post = ["foto", "video", "reel", "carrusel", "historia"]
```

```python
def fecha_aleatoria():
    inicio = datetime(2023, 1, 1)
    dias = random.randint(0, 730)
    return (inicio + timedelta(days=dias)).strftime("%Y-%m-%d")
```

```python
cantidad = 1_000_000
```

```python
dfInstagram = pd.DataFrame({
    "influencer":   [random.choice(influencers) for _ in range(cantidad)],
    "likes":        [random.randint(100, 500_000) for _ in range(cantidad)],
    "comentarios":  [random.randint(10, 50_000) for _ in range(cantidad)],
    "fecha":        [fecha_aleatoria() for _ in range(cantidad)],
    "tipo_post":    [random.choice(tipos_post) for _ in range(cantidad)],
})
```

```python
print(dfInstagram.shape)
dfInstagram.head(10)
```

```python
# Promedio de likes y comentarios por influencer
engagement = dfInstagram.groupby("influencer").agg(
    likes_promedio    = ("likes", "mean"),
    comentarios_total = ("comentarios", "sum"),
    total_posts       = ("influencer", "count")
).sort_values("likes_promedio", ascending=False)

print(engagement)
```

```python
# Tipo de post más popular
print(dfInstagram["tipo_post"].value_counts())
```
